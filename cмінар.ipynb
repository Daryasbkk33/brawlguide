## №2 Телефон (1б)

Услуги телефонной сети оплачиваются по следующему правилу: за разговоры до А минут в месяц – В рублей за минуту, а разговоры сверх установленной нормы оплачиваются из расчета С рублей за минуту. Требуется написать программу, вычисляющую плату за пользование телефоном для разговоров продолжительностью Т минут в месяц.

**Входные данные**

Первая строка содержит натуральные числа через пробел A, B, C и T, не превышающие 1000.

**Выходные данные**

Выведите единственное целое число – месячную плату за пользование телефоном.

| input | output|
|:-------:|:------:|
| 30 2 3 30     | 60  |
| 20 1 3 30   | 50    |

https://acmp.ru/index.asp?main=task&id_task=933

## №5 Сбор земляники   (1б)

Маша и Миша собирали землянику. Маше удалось сорвать X ягод, а Мише – Y ягод. Поскольку ягода была очень вкусной, то ребята могли какую-то часть ягод съесть. По нашим подсчетам вместе они съели Z ягод.

Требуется определить: сколько ягод ребята собрали в результате, при этом следует проверить, не ошиблись ли мы в расчетах, подсчитывая количество съеденных ягод (их не должно было получиться больше, чем сорванных ягод).

**Входные данные**

Строка содержит три натуральных числа X, Y и Z, не превышающих 1000. Все числа расположены в первой строке файла и разделены пробелом.

**Выходные данные**

Выведите количество собранных ягод, если наши подсчеты оказались правдоподобными, либо слово «Impossible» в противном случае.


| input | output|
|:-------:|:------:|
| 3 2 1| 4|
| 12 13 5| 20|
| 2 5 9 | Impossible|

https://acmp.ru/index.asp?main=task&id_task=755

```python
#code
X, Y, Z = map(int, input().split())
if Z > X + Y:
  print ('Impossible')
else:
  print (X + Y - Z)
```

## №6 Мышка   (1б)

Однажды компьютерная мышка подумала, что стоит взять про запас еще один коврик. Чтобы никто не заметил запасного коврика, мышка решила его спрятать под свой прямоугольный коврик. Пробравшись ночью на склад, мышка обнаружила там только круглые коврики. Удастся ли мышке спрятать круглый коврик под прямоугольным ковриком?

**Входные данные**

Первая строка содержит три натуральных числа через пробел: W, H и R, где W и H - ширина и высота прямоугольного коврика, а R – радиус запасного коврика.

**Выходные данные**

Выведите «YES», если новый коврик можно спрятать под старым, и слово «NO», если этого сделать нельзя.


| input | output|
|:-------:|:------:|
| 4 7 2| YES|
| 4 7 3| NO|

https://acmp.ru/index.asp?main=task&id_task=907

```python
#code
W, H, R = map(int, input().split())
if 2*R < H and 2*R < W:
  print('YES')
else:
  print('NO')
```

## №9 От перестановки что-то меняется ...

Всем известно, что «от перестановки слагаемых сумма не изменяется». Однако, случается и так, что перестановка двух чисел приводит к более интересным последствиям.

Пусть, например, заданы три числа: a1, a2, a3. Рассмотрим равенство a1+ a2= a3. Оно может быть неверным (например, если a1= 1, a2= 4, a3= 3), однако может стать верным, если поменять некоторые числа местами (например, если поменять местами a2 и a3, оно обратится в равенство 1 + 3 = 4).

Ваша задача – по заданным трем числам определить: можно ли их переставить так, чтобы сумма первых двух равнялась третьему.

**Входные данные**

Первая строка содержит три целых числа через пробел: a1, a2, a3 (−108 ≤ a1, a2, a3 ≤ 108).

***Выходные данные**

Выведите слово «YES», если заданные числа можно переставить так, чтобы сумма первых двух равнялась третьему. В противном случае выведите в выходной файл слово «NO».

| input | output|
|:-------:|:------:|
| 3 5 2 | YES|
| 2 2 5| NO|
| 2 2 4| YES|

https://acmp.ru/index.asp?main=task&id_task=970

```python
#code
N, A, B = list(map(int, input().split()))
if A + B == N or A + N == B or B + N == A:
  print('YES')
else:
  print('NO')
```

## №10 Клавиатура

Для данной буквы английского алфавита нужно вывести справа стоящую букву на стандартной клавиатуре. При этом клавиатура замкнута, т.е. справа от буквы «p» стоит буква «a», от буквы «l» стоит буква «z», а от буквы «m» — буква «q».

**Входные данные**

Первая строка содержит один символ — маленькую букву английского алфавита.

**Выходные данные**

Cледует вывести букву стоящую справа от заданной буквы, с учетом замкнутости клавиатуры

| input | output|
|:-------:|:------:|
| q | w|
| w| e|
| t| y|
| m| q|
| l| z|

https://acmp.ru/index.asp?main=task&id_task=66

```python
# code
alpha = 'qwertyuiopasdfghjklzxcvbnm'
omega = input().strip()
index = alpha.index(omega)
sigma = alpha[(index + 1) % len(alpha)]
print(sigma)
```
